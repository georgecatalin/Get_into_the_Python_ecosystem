## Classes in Python

## using static members of a class in a objects 

```python
class Dog:
    _legs = 4
    def __init__(self,name):
        self.name = name

    def getLegs(self):
        return self._legs

    def bark(self):
        return('{0} barks!'.format(self.name))

mydog = Dog('Zizou')
print(mydog.getLegs())
print(mydog.bark())
```

## using static members of a class as properties of the class

```python
class Dog:
    _legs = 4
    def __init__(self,name):
        self.name = name;

    def getLegs():
        return Dog._legs

    def bark(self):
        return(f'{self.name} is barking!')

mydog = Dog('Zizan')
print(Dog.getLegs())
print(mydog.bark())
```

## static and instance methods


```python
class Wordset:
    def __init__(self):
        self.wordset = set()

    def addText(self,text):
        text = Wordset.clearText(text)
        for word in text.split():
            self.wordset.add(word)

    def clearText(text):
        #chaining functions
        text = text.replace('!','').replace('\'','').replace('.','')
        return text.lower()

wordset = Wordset();
wordset.addText('Hello I\'am George Calin!')
wordset.addText('And I am 47 years old')

print(wordset.wordset)
```

### rewrite the class with a static member

```python
class Wordset:
    _replace_punctuation = ('!','.',',','\'')
    def __init__(self):
        self.wordset = set()

    def addText(self,text):
        text = Wordset.clearText(text)
        for word in text.split():
            self.wordset.add(word)

    def clearText(text):
        for punctuation in Wordset._replace_punctuation:
            text = text.replace(punctuation,'')
        return text.lower()

wordset = Wordset();
wordset.addText('Hello I\'am George Calin!')
wordset.addText('And I am 47 years old')

print(wordset.wordset)
```

### writing the same code with a decorator

```python
class Wordset:
    _replace_punctuation = ('!','.',',','\'')
    def __init__(self):
        self.wordset = set()

    def addText(self,text):
        text = self.clearText(text)   #modified here
        for word in text.split():
            self.wordset.add(word)

    @staticmethod    #this is the decorator
    def clearText(text):
        for punctuation in Wordset._replace_punctuation:
            text = text.replace(punctuation,'')
        return text.lower()

wordset = Wordset();
wordset.addText('Hello I\'am George Calin!')
wordset.addText('And I am 47 years old')

print(wordset.wordset)
```
