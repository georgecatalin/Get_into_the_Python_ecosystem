# Threads

```python
import threading
import time
```

```python
def calculateLongSquare(number):
    time.sleep(1)
    return number ** 2

[calculateLongSquare(number) for number in range(0,10)]
```

### Rewrite with threads

```python
def calculateSquare(number,result):
    time.sleep(1)
    result[number]=number ** 3

result=dict();
thread1 = threading.Thread(target=calculateSquare,args=(1,result))
thread2 = threading.Thread(target=calculateSquare,args=(2,result))

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print(result)
```

#### Rewrite with threads in a much better way

```python
def calculateRoot(number,results):
    time.sleep(1)
    results[number]=number**0.5

results = {}

threads = [threading.Thread(target=calculateRoot,args=(n,results)) for n in range(0,200)]
[t.start() for t in threads]
[t.join() for t in threads]

# [print(f'{key}:{value}') for key,value in results.items()]
for key,value in results.items():
    print('{0}:{1}'.format(key,value))
    
```
