## Errors and exceptions

```python
import time
```

```python
def function_that_raises_error():
    return 1/0

function_that_raises_error()
```

### Try and except

```python
try:
    function_that_raises_error()
except Exception as e:
    print(type(e))
```

#### Handling exceptions

```python
def cause_error():
    try:
        return 1/0
    except Exception:
        print('There was an error')

cause_error()
```

### Finally

```python
def cause_error_version_timer():
    start_time = time.time()
    try:
        return 1/0
    except Exception:
        print('Some sort of error occurred')
    finally:
        print(f'The execution of this code took {time.time()-start_time} seconds')

cause_error_version_timer()
```

### Catch exceptions by their type

```python
def cause_error_version_2():
    try:
        return 1/0
    except TypeError:
        print('There was a type error')
    except ZeroDivisionError:
        print('There was a division by zero error')
    except Exception:
        print('There was some other general kind of error')
    finally:
        print('This portion of the code will be running all the time')

cause_error_version_2()
```

### Use decorators to handle exceptions 

```python
def handle_exceptions(function):
    def wrapper(*args):
        try:
            function(*args)
        except TypeError:
            print('There was a type error')
        except ZeroDivisionError:
            print('There was a division by zero error')
        except Exception:
            print('There was some other general kind of error')
    return wrapper

@handle_exceptions
def raise_error():
    return 1/0

raise_error()
```

### Raising Exceptions

```python
@handle_exceptions
def raise_exception(n):
    if n == 0:
        raise Exception()
    print(n)

raise_exception(1)
```

## Custom Exceptions

```python
class CustomException(Exception):
    pass

def raise_custom_error():
    raise CustomException('You got into serious troubles')

raise_custom_error()
```
