## Inheritance

```python
class Dog:
    _legs = 4
    def __init__(self,name):
        self.name = name

    def bark(self):
        print('{0} says Bark!!!'.format(self.name))

    def getLegs(self):
        return self._legs

class ShihTzu(Dog):
    def bark(self):
        print(f'{self.name} says HowHow!!')

zizou = ShihTzu('Zizou')
zizou.bark()

milla = Dog('Milluca')
milla.bark()
```

## Extend built-in Python classes

```python
class UniqueList(list):
    def __append__(self,item):
        if item in self:
            return
        super().append(item)

unique = UniqueList()
unique.append(1978)
unique.append(2011)
unique.append(1977)
unique.append(1978)

print(unique)
```

#### Modify the constructor of the Child class

```python
class Unique(list):
    def __init__(self):
        super().__init__()
        self.specialMember = 'I am special'

    def append(self, item):
        if item in self:
            return
        super().append(item)

unique = Unique()
unique.append(1953)
unique.append(1954)
unique.append(1978)
unique.append(1978)

print(unique.specialMember)
print(unique)
```
